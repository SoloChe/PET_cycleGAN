```python
import pandas as pd
import numpy as np

uFBP_sep_path = '/data/amciilab/processedDataset/ADNI/ADNI-PUP/results/FBP/list_ADNI_FBP_SUVRLR.csv'
uPiB1_sep_path = '/data/amciilab/processedDataset/OASIS/Oasis-PUP/results/PiB_IDS_SUVRLR.csv'
uPiB2_sep_path = '/data/amciilab/processedDataset/AIBL/AIBL-PUP/results/PiB_IDS_SUVRLR.csv'

pFBP_sep_path = '/data/amciilab/processedDataset/Centiloid/FBP_PIB/PUP_FBP/runExtract/list_id_SUVRLR.csv'
pPiB_sep_path = '/data/amciilab/processedDataset/Centiloid/FBP_PIB/PUP_PIB/runExtract/list_id_SUVRLR.csv'

uFBP_sep = pd.read_csv(uFBP_sep_path) 
uPiB1_sep = pd.read_csv(uPiB1_sep_path)
uPiB2_sep = pd.read_csv(uPiB2_sep_path)
uPiB_sep = pd.concat([uPiB1_sep, uPiB2_sep])

pFBP_sep = pd.read_csv(pFBP_sep_path)
pPiB_sep = pd.read_csv(pPiB_sep_path)

uPiB_sep.head()
```

```python
# check if the columns are all 1 or 0
names = []
for i, name in enumerate(uFBP_sep.columns):
    if len(uFBP_sep[name].unique()) == 1: 
        names.append(name)
        
names = names + ['Left-Cerebellum-Cortex', 'Right-Cerebellum-Cortex']  
print(names)
```

```python
# drop the columns
uFBP_sep = uFBP_sep.drop(columns=names)
uPiB_sep = uPiB_sep.drop(columns=names)
pFBP_sep = pFBP_sep.drop(columns=names)
pPiB_sep = pPiB_sep.drop(columns=names)

print(uFBP_sep.shape, uPiB_sep.shape, pFBP_sep.shape, pPiB_sep.shape)
```

```python
for i in uFBP_sep.columns: 
    print(i)
```

```python
region_index = ['ctx-precuneus', 
                'ctx-rostralmiddlefrontal', 
                'ctx-superiorfrontal', 
                'ctx-middletemporal', 
                'ctx-superiortemporal', 
                'ctx-lateralorbitofrontal', 
                'ctx-medialorbitofrontal']

lh = {}
rh = {}
for region in region_index: 
    newr = 'ctx-rh-' + region[4:]
    newl = 'ctx-lh-' + region[4:]
    
    idxr = np.where(uFBP_sep.columns == newr)[0]
    idxl = np.where(uFBP_sep.columns == newl)[0]
    rh[newr] = idxr[0]-1
    lh[newl] = idxl[0]-1

```
